```python
#%% import modules 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from scipy.interpolate import griddata
# import h5py 
plt.rcParams.update({'font.size': 18})
```

```python
melt = pd.read_csv('/home/lxue07/Documents/ThermoMech2d/ForLiang_ThermoMech2D/\
ForLiang_ThermoMech2D/output/AfricaModels2022/Table/\
Lake_drop_model_restart_69_1000m_melting_history_123-191_dt_1_nomark_36.csv')
melt.head()

melt_nochange = pd.read_csv('/home/lxue07/Documents/ThermoMech2d/ForLiang_ThermoMech2D/\
ForLiang_ThermoMech2D/output/AfricaModels2022/Table/\
Lake_nochange_model_restart_69_1000m_melting_history_123-191_dt_1_nomark_36.csv')

melt.describe()
```

```python
# peridotite  components 
Olivine= 0.8
Opx = 0.1
Cpx = 0.5
Gar = 0.5
#t = 9
ree = pd.read_csv('../table/Kelemen_2002.csv', sep = '\t')
ree.head()
```

```python


# fraction 
step = 45
mf_drop= melt['max_mf'].iloc[step]
mf_nochange = melt_nochange['max_mf'].iloc[step]
# distribution coefficient
ree['D'] = Olivine* ree['ol/liq']+ Opx*ree['opx/liq'] + Cpx*ree['cpx/liq'] +Gar*ree['gar/liq']

# Cl/Co = 1/(F+D-F*D)
#ree= ree.iloc[:-2]
ree['Cl/Co_drop'] = 1/(mf_drop + ree['D'] - ree['D']*mf_drop)
ree['Cl/Co_nochange'] = 1/(mf_nochange + ree['D'] - ree['D']*mf_nochange)

# ree['Cl/Co_l_normalized'] =ree['Cl/Co_l']/ree['chondrite(ppm)']
# ree['Cl/Co_ul_normalized'] =ree['Cl/Co_ul']/ree['chondrite(ppm)']

# ree['Cl/Co_change in_per'] =  (ree['Cl/Co_ul']- ree['Cl/Co_l'])/ree['Cl/Co_l']
ree.head()
```

```python
fig, ax = plt.subplots(figsize=(9,5))

plt.scatter(ree['element'], ree['Cl/Co_drop'])
plt.plot(ree['element'], ree['Cl/Co_drop'], label = 'REE from lake level drop')
plt.scatter(ree['element'], ree['Cl/Co_nochange'])
plt.plot(ree['element'], ree['Cl/Co_nochange'], label = 'REE from constant lake level')

# plt.scatter(ree['element'], ree['Cl/Co_l_normalized'])
# plt.plot(ree['element'], ree['Cl/Co_l_normalized'], label = 'REE from lake level drop')
# plt.scatter(ree['element'], ree['Cl/Co_ul_normalized'])
# plt.plot(ree['element'], ree['Cl/Co_ul_normalized'], label = 'REE from constant lake level')

# plt.scatter(ree['element'], ree['Cl/Co_change in_per']*100)
# plt.plot(ree['element'], ree['Cl/Co_change in_per']*100, 
#          label = 'REE concentration lake level drop vs constant lake level ')

plt.legend(loc = 'lower left', fontsize = 15)
plt.grid(linestyle='dotted')
plt.tight_layout()
#plt.yscale('log')
#plt.ylabel('concentration normalized to chondrites')
plt.ylabel('REE changes in %')
plt.ylabel('Cl/Co')
# note that Ba is between Rb and Th 
```

```python
#Ba Cl/Co
nb_r = ree['Cl/Co_drop'].iloc[4]
#Ba in V km3

# CO2 ppm vs Rb, Nb, Ba ppm are linear 
# CO2/Nb = 239- 557 Le Voyer et al., 2017 nature communications
# From Zong et al., 2020,   Rb 49.81 ppm , 
co2_c = nb_r*49.81*557

# ppm to mg/l: 1 ppmapproximately 1 mg/L
# from v to mg
# melt['melt_v'] is in km2/km to dm2/km
co2_mg = co2_c* melt['melt_v'].iloc[step] * (1e3)**2
co2_ton = co2_mg*1e-9
co2_ton
```

```python
# here we only need Nb, so choose ree['D'].iloc[4]
melt['Nb_Cl/Co_drop'] = 1/(melt['max_mf'] + ree['D'].iloc[4] - ree['D'].iloc[4]*melt['max_mf'])
melt['Nb_Cl/Co_nochange'] = 1/(melt_nochange['max_mf'] + ree['D'].iloc[4] 
                               - ree['D'].iloc[4]*melt_nochange['max_mf'])
melt['co2_ton_drop'] = melt['Nb_Cl/Co_drop']* 49.81*557 *melt['melt_v']* (1e3)**2 *1e-9
melt['co2_ton_nochange'] = melt['Nb_Cl/Co_nochange']* 49.81*557* melt_nochange['melt_v']* (1e3)**2 *1e-9
melt['co2_ton_drop_low'] = melt['Nb_Cl/Co_drop']* 49.81*239 *melt['melt_v']* (1e3)**2 *1e-9
melt['co2_ton_nochange_low'] = melt['Nb_Cl/Co_nochange']* 49.81*239* melt_nochange['melt_v']* (1e3)**2 *1e-9
melt.head()
```

```python
melt['time_ky'] = melt['time_ky']*1e-3
```

```python
#co2 degassing:
#Ba: ol/liq: 1.000000e-09 ppm milligrams per liter (mg/L) = 1e9 g/km3


#ree = pd.read_csv('../table/Kelemen_2002.csv', sep = '\t')
#ree.head()

fig, ax = plt.subplots(figsize=(12,7))
ln1 = ax.plot(melt['time_ky']- melt['time_ky'][0], melt['water_mass_kg'], label = 'surface water mass', c = 'C1')
ax.set_xlabel('time (ky)')
ax.set_ylabel('Water mass (kg/km)')
ax2 = ax.twinx()
ln2 = ax2.plot(melt['time_ky']- melt['time_ky'][0], melt['co2_ton_drop'] - melt['co2_ton_nochange'],  
               c = 'C0',label = 'CO2 difference high end member')
ln3 = ax2.plot(melt['time_ky']- melt['time_ky'][0], melt['co2_ton_drop_low'] - melt['co2_ton_nochange_low'], '--',
               c = 'C0',label = 'CO2 difference low end member')

ax2.set_ylabel('differential CO2 flux (ton/km)')
lns = ln1 +ln2 + ln3
labs = [l.get_label() for l in lns]
ax2.legend(lns, labs, loc = 'lower left')
plt.grid(linestyle='dotted')
plt.tight_layout()
plt.xlim([0.05, 0.27])
plt.savefig('../Figure/co2_t.jpg')
```
