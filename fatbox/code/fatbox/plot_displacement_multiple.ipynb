```python
length_ls = []
for time in list(range(len(Gs))):
    sum_dis = 0
    for n in Gs[time].edges:
        #print (sum_dis)
        value = Gs[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value #into kyr
    length_ls.append(sum_dis)

length_ls_r = []
for time in list(range(len(Gs_r))):
    sum_dis = 0
    for n in Gs_r[time].edges:
        #print (sum_dis)
        value = Gs_r[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value #into kym
    length_ls_r.append(sum_dis)    
#plt.scatter(range(69), length_ls)
plt.plot(t, length_ls, label = 'drop')
plt.plot(t, length_ls_r, label = 'ref')
plt.legend()
#plt.plot(list(np.linspace(1,70*5, 69)), length_ls)  
```
