```python
t = [x*5 for x in list (range(len(Gs)))]
ll = np.zeros(len(t)) # lake level
print (max(t))
period = 20 # index
for i in range(len(t)):
    if i>= 15 and i<= 55:
        ll[i] = -400-400*np.sin(i*2*np.pi/period) 
    if i<15 or i>55:
        ll[i] = 0
plt.plot(t,ll)
```

```python
dis_ls = []
for time in list(range(len(Gs))):
    sum_dis = 0
    for n in Gs[time].nodes:
        #print (sum_dis)
        value = Gs[time].nodes[n]['displacement']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*100 #into kyr
    dis_ls.append(sum_dis)
    
dis_ls_r = []    
for time in list(range(len(Gs_r))):
    sum_dis = 0
    for n in Gs_r[time].nodes:
        #print (sum_dis)
        value = Gs_r[time].nodes[n]['displacement']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*100 #into kyr
    dis_ls_r.append(sum_dis)
    
plt.plot(t, dis_ls)
#plt.plot(time_kyr, dis_ls_r)
```

```python
length_ls = []
for time in list(range(len(Gs))):
    sum_dis = 0
    for n in Gs[time].edges:
        #print (sum_dis)
        value = Gs[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*100 #into kyr
    length_ls.append(sum_dis)

length_ls_r = []
for time in list(range(len(Gs_r))):
    sum_dis = 0
    for n in Gs_r[time].edges:
        #print (sum_dis)
        value = Gs_r[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*100 #into kyr
    length_ls_r.append(sum_dis)    
#plt.scatter(range(69), length_ls)
plt.plot(t, length_ls)  
```

```python
# extract x and length data
length_ls = []
cen_x_ls = []
for time in list(range(len(Gs))):
    leng = []
    cen_x = []
    for n in Gs[time].edges:
        #print (sum_dis)
        x1= Gs[time].nodes[n[0]]['pos'][0]
        y1= Gs[time].nodes[n[0]]['pos'][1]
        x2= Gs[time].nodes[n[1]]['pos'][0]
        y2= Gs[time].nodes[n[1]]['pos'][1]
        value = Gs[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        leng.append(value*100) #into m
        cen_x.append(0.5*(x1+x2)*100) # into m
    length_ls.append(leng)
    cen_x_ls.append(cen_x)
    
# prepare project x range 
x_min = 0
if fault_type ==1:
    x_max = 55e3
if fault_type ==2:
    x_max = 50e3
    
n_interval = 100
mark_ls = np.linspace(x_min, x_max, n_interval)    
    
# project x range
length_xt = []
for t in range(len(cen_x_ls)):
    sum_l = [] # length in each time step 
    for j in range(len(mark_ls)-1):
        index = np.where(np.logical_and(cen_x_ls[t]>=mark_ls[j], cen_x_ls[t]<=mark_ls[j+1]))
        sum_l_x = 0 # lengh in x
        for i in index[0]:
            #print(i)
            sum_l_x = sum_l_x + length_ls[t][i]
        sum_l.append(sum_l_x)
    length_xt.append(sum_l)
    
# plot figure    
fig, ax = plt.subplots(figsize = (6,10))
if fault_type ==1:
    plt.imshow(length_xt, aspect=0.4, cmap ='RdYlBu_r', vmax = 3000, extent=[105,160,350,0])
if fault_type ==2:
    plt.imshow(length_xt, aspect=0.4, cmap ='RdYlBu_r', vmax = 3000, extent=[15,55,350,0])

cbar = plt.colorbar(label = 'Fault length (m)')
plt.xlabel('Horizontal distance (km)')
plt.ylabel('Model time (kyr)')
plt.tight_layout()
#plt.savefig('../../Figure/restart_71_secondary_ft_migration.jpg')
#plt.savefig('../../Figure/restart_71_primary_ft_migration.jpg')
```

```python
# extract x and displacement data
dis_ls = []
x_ls = []
for time in list(range(len(Gs))):
    dis = []
    x = []
    for n in Gs[time].nodes:
        #print (sum_dis)
        x1= Gs[time].nodes[n]['pos'][0]
        value = Gs[time].nodes[n]['slip_x']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        dis.append(value*100) #into m
        x.append(x1*100) # into m
    dis_ls.append(dis)
    x_ls.append(x)
# prepare project x range 
x_min = 0
if fault_type ==1:
    x_max = 55e3
if fault_type ==2:
    x_max = 50e3
    
n_interval = 100
mark_ls = np.linspace(x_min, x_max, n_interval)    
    
# project x range
dis_xt = []
for t in range(len(x_ls)):
    sum_l = [] # length in each time step 
    for j in range(len(mark_ls)-1):
        index = np.where(np.logical_and(x_ls[t]>=mark_ls[j], x_ls[t]<=mark_ls[j+1]))
        sum_l_x = 0 # lengh in x
        for i in index[0]:
            #print(i)
            sum_l_x = sum_l_x + dis_ls[t][i]
        sum_l.append(sum_l_x)
    dis_xt.append(sum_l)
    
# # plot figure    
fig, ax = plt.subplots(figsize = (6,10))
if fault_type ==1:
    plt.imshow(dis_xt, aspect=0.4, cmap ='RdYlBu_r',vmax =1e-8, extent=[105,160,350,0])
if fault_type ==2:
    plt.imshow(dis_xt, aspect=0.4, cmap ='RdYlBu_r',extent=[15,55,350,0])

cbar = plt.colorbar(label = 'displacement (m)')
plt.xlabel('Horizontal distance (km)')
plt.ylabel('Model time (kyr)')
plt.tight_layout()
# plt.savefig('../../Figure/restart_71_secondary_ft_migration.jpg')
# #plt.savefig('../../Figure/restart_71_primary_ft_migration.jpg')
```
