```python
t = [x*5 for x in list (range(len(Gs)))]
ll = np.zeros(len(t)) # lake level
print (max(t))
period = 20 # index
for i in range(len(t)):
    if i>= 15 and i<= 55:
        ll[i] = -400-400*np.sin(i*2*np.pi/period) 
    if i<15 or i>55:
        ll[i] = 0
plt.plot(t,ll)
```

```python
#for t in in list(range(len(Gs))):
no_ft = []
for time in list(range(len(Gs))):
    max_ft = 0    
    for n in Gs[time].nodes:
        value = Gs[time].nodes[n]['strain_rate']
        if value > max_ft:
            max_ft = value
    no_ft.append(max_ft)
    
no_ft_r = []
for time in list(range(len(Gs))):
    max_ft = 0    
    for n in Gs_r[time].nodes:
        value = Gs_r[time].nodes[n]['strain_rate']
        if value > max_ft:
            max_ft = value
    no_ft_r.append(max_ft)    
plt.plot(t, no_ft) 
plt.plot(t, no_ft_r)
#plt.plot(t, [x-y for x,y in zip(no_ft, no_ft_r)])
```

```python
#for t in in list(range(len(Gs))):
no_ft = []
for time in list(range(len(Gs))):
    max_ft = 0    
    for n in Gs[time].nodes:
        value = Gs[time].nodes[n]['fault']
        if value > max_ft:
            max_ft = value
    no_ft.append(max_ft)
    
no_ft_r = []
for time in list(range(len(Gs))):
    max_ft = 0    
    for n in Gs_r[time].nodes:
        value = Gs_r[time].nodes[n]['fault']
        if value > max_ft:
            max_ft = value
    no_ft_r.append(max_ft)    
plt.plot(t, no_ft) 
plt.plot(t, no_ft_r)
#plt.plot(t, [x-y for x,y in zip(no_ft, no_ft_r)])
```

```python
dis_ls = []
for time in list(range(len(Gs))):
    sum_dis = 0
    for n in Gs[time].nodes:
        #print (sum_dis)
        value = Gs[time].nodes[n]['displacement']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*60*60*2*365.25*1000 #into kyr
    dis_ls.append(sum_dis)
    
dis_ls_r = []    
for time in list(range(len(Gs_r))):
    sum_dis = 0
    for n in Gs_r[time].nodes:
        #print (sum_dis)
        value = Gs_r[time].nodes[n]['displacement']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value*60*60*2*365.25*1000 #into kyr
    dis_ls_r.append(sum_dis)
    
plt.plot(t, dis_ls)
plt.plot(t, dis_ls_r)
```

```python
# accumulative discplacement 
dis_acc = []
j = 0
for i in range (len(dis_ls)):
    j+=dis_ls[i]
    dis_acc.append(j)
    
dis_acc_r = []
j = 0
for i in range (len(dis_ls_r)):
    j+=dis_ls_r[i]
    dis_acc_r.append(j)
    
    
plt.plot(t, dis_acc)
plt.plot(t, dis_acc_r)
```

```python
length_ls = []
for time in list(range(len(Gs))):
    sum_dis = 0
    for n in Gs[time].edges:
        #print (sum_dis)
        value = Gs[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value
    length_ls.append(sum_dis)

length_ls_r = []
for time in list(range(len(Gs_r))):
    sum_dis = 0
    for n in Gs_r[time].edges:
        #print (sum_dis)
        value = Gs_r[time].edges[n]['length']
        if math.isnan(value):
            #print('---nan---')
            value = 0
        sum_dis += value 
    length_ls_r.append(sum_dis)    
#plt.scatter(range(69), length_ls)
plt.plot(list(np.linspace(1,70*5, 69)), length_ls)  
```
