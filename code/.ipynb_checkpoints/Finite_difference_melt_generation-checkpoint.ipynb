```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import numpy as np
%matplotlib inline
plt.rcParams.update({'font.size': 15})
```

```python
df = pd.read_csv('../AfricaModels2022/Table/Lake_drop_restart_71_1000m_depth_meltf_FD_plot_135_to_203_dt2depthf20.txt')
df = pd.read_csv('../AfricaModels2022/Table/Lake_drop_restart_71_1000m_depth_meltf_FD_plot_135_to_203_dt2depthf20.txt')

df.describe()
# time: kyr
# depth:km
# mf_diff:%
```

```python
#t = [step_b: int: step_e];
t= np.arange (135, 203+1, 2).tolist()
#  depth = linspace(20, maxdepth, nfreq+1);
depth = np.linspace(0, 100, 20+1)
print (depth)
```

```python

xv = np.linspace(min(t), max(t), len(t));
yv = np.linspace(min(depth), max(depth), len(depth));
[Xm,Ym] = np.meshgrid(xv, yv);
Zm = griddata((df['time'], df['depth']), df['mf_diff'], (Xm, Ym), method = 'cubic');
dt = 2*5e3; # time difference in year
```

```python
plt.imshow(Zm)
```

```python
Z_rate = np.zeros([Zm.shape[0], Zm.shape[1]]) ;

for i in range(0,Zm.shape[1]-1):
    #print (i)# x
    for j in range(0,Zm.shape[0]-1):
        #print ('y', j)# y
        Z_rate[j,i] = (Zm[j,i+1]-Zm[j,i])/(1*dt)
```

```python
plt.imshow(Z_rate)
plt.colorbar()
```

```python
#  two lake level lows: 110kyr 220 kyr
fig, ax = plt.subplots(figsize=(12, 5))
plt.contourf([(x-min(t))*5 for x in t],depth, Z_rate, cmap= 'RdYlBu_r')
#D
plt.colorbar(label=" Difference in melt production rate \n(kg $m^{-3}$ $yr^{-1}$)")
ax.invert_yaxis()
ax.set_ylabel('Depth (km)')
ax.set_xlabel(' Time (kyr)')
plt.ylim([70, 0])
plt.tight_layout()
plt.savefig('../AfricaModels2022/Figure/extraction/melt_production1.jpg')
```

```python
import scipy as sp
import scipy.ndimage
sigma = [1,1]
Z_s = sp.ndimage.filters.gaussian_filter(Z_rate, sigma, mode='constant')

#  two lake level lows: 110kyr 220 kyr
fig, ax = plt.subplots(figsize=(12, 5))
plt.contourf([(x-min(t))*5 for x in t],depth, Z_s, cmap= 'RdYlBu_r')
#D
plt.colorbar(label=" Difference in melt production rate \n(kg $m^{-3}$ $yr^{-1}$)")
ax.invert_yaxis()
ax.set_ylabel('Depth (km)')
ax.set_xlabel(' Time (kyr)')
plt.ylim([70, 0])
plt.tight_layout()
#plt.savefig('../AfricaModels2022/Figure/extraction/melt_production1_smooth.jpg')
```
